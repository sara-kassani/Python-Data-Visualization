```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8,8))
plt.rcParams.update({'font.size': 16})

labels = 'Dysplasia and Cancer', 'Gastric Metaplasia', 'Intestinal Metaplasia', 'Squamous Epithelium'
sizes = [3594, 1206, 3177, 1469]



plt.pie(sizes,  labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
plt.axis('equal')
# plt.legend(loc='lower right', bbox_to_anchor=(1.0, 1.00))
plt.savefig('pie_chart.png', bbox_inches='tight', dpi = 100)

plt.show()
```
